```python
class GR:
    def __init__(self, learning_rate, epochs):
        self.m =-100
        self.b = 100
        self.lr = learning_rate
        self.epochs = epochs
        
    def fit(self,X,Y):
        for i in range(self.epochs):
            loss_function_b = -2*np.sum(Y- self.m*X.ravel()-self.b)  
            loss_function_m = -2*np.sum((Y- self.m*X.ravel()-self.b)*X.ravel())
            
            self.b = self.b - (self.lr*loss_function_b)
            self.m = self.m - (self.lr*loss_function_m)
            y_pred = self.m*X.ravel()+self.b
            plt.plot(X,y_pred)
            print(self.m,self.b)
        plt.plot(X,reg.predict(X), color = 'black') # this is the line of the linear regression function by scikit learn
        print(self.m,self.b)
    def predict(self,X_test):
        return self.m*X_test.ravel() + self.b
```

```python
gd = GR(0.01,10)
```

```python
gd.fit(X_train,y_train)
```

```python
y_predict = gd.predict(X_test)
```
