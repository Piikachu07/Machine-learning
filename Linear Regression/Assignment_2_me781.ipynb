```python
import numpy as np
```

```python
N = 100
m = 140
X_vector  = np.random.normal(loc = 0,scale =np.sqrt(1/N), size =(m,N))
```

```python
X_vector[10]
```

```python
distances = []
for i in X_vector:
    distance = np.linalg.norm(X_vector)
    distances.append(distance)
    
```

```python
len(distances)
```

```python
distances
```

```python
angle_in_degrees = []
for i in range(0,len(X_vector)-1):
    dot_product = np.dot(X_vector[i],X_vector[i+1])
    cos_thetha = dot_product / (distances[i]*distances[i+1])
    angle_in_radians = np.arccos(cos_thetha)
    angle_in_degree = np.degrees(angle_in_radians)
    angle_in_degrees.append(angle_in_degree)
```
